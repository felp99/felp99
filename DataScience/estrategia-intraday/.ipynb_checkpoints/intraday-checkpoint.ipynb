# Estratégia **Intraday** | A verdadeira forma de ganhar dinheiro com ações dormindo

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
ticker = 'BOVA11.SA'
period = '3y'
interval = '1d'

closed = yf.Ticker(ticker).history(interval = '1d', period=period)[['Open','Close']]
```

### Algoritmo para criação das colunas

```python
df = pd.DataFrame(closed)
investment = 10

df['IntradayProfit'] = (df['Open'].shift(-1) - df['Close']).shift() * investment
df['DayProfit'] = (df['Close'] - df['Open']).shift() * investment
```

### Resultado parcial

```python
fig = df[['IntradayProfit', 'DayProfit']].cumsum().plot(figsize = (14,8), title= f'{interval} {ticker} Analysis')
fig.title.set_size(20)
```

### Forma de armazenamento dos dados

```python
p_result = pd.DataFrame(columns = [ticker]).transpose()
p_result[['IntradayProfit', 'DayProfit']] = df[['IntradayProfit', 'DayProfit']].iloc[-1], 
p_result
```

### Ampliação e comprovação dos resultados

```python
result = pd.DataFrame()

for ticker in (tickers):
    
    df = pd.read_csv(f'data/{ticker}')

    df['IntradayProfit'] = (df['Open'].shift(-1) - df['Close']).shift() 
    df['DayProfit'] = (df['Close'] - df['Open']).shift() 
    
    result.loc[ticker, ['IntradayProfit', 'DayProfit']] = df[['IntradayProfit', 'DayProfit']].iloc[-1]
result.dropna(how='any', inplace=True)
```
