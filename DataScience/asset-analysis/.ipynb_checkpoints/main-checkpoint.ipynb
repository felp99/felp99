```python
import gspread
import datetime
import requests
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

CODE = '1LyxXkYYPWU_UkuydQhpPWlc-DAM2oMYXKrX4AryEkH8'

gc = gspread.service_account(filename = '../key.json')
sh = gc.open_by_key(CODE)

ws = sh.worksheet('Extrato Investimentos') #Lista dos WS

_values = ws.get_values(value_render_option='UNFORMATTED_VALUE')
```

```python
df_values = pd.DataFrame(_values[1:], columns = _values[0])
df_values.drop(columns=['ID'], inplace=True)

time_init = datetime.datetime(1899, 12, 30)
timedelta = pd.to_timedelta(df_values.Data, 'd')

df_values.loc[:, 'Data'] = time_init + timedelta
df_values.drop(columns=['Preço Médio'], inplace=True)
```

```python
dict_transform = {
    'ITSA3': 'ITSA3',
    'BBDC4': 'BBDC4',
    'PICPAY': 'PICPAY',
    'BITCOIN': 'BTC',
    'ETHEREUM': 'ETH',
    'USDCOIN': 'USDC',
    'PAX GOLD': 'PAXG',
    'XRP': 'XRP',
    'LITECOIN': 'LTC',
    'CARDANO': 'ADA',
    'HEDERA': 'HBAR',
    'ZRX': 'ZRX',
    'MANA': 'MANA',
    'SUSHI': 'SUSHI',
    'SAND': 'SAND',
    'SOLANA': 'SOL',
    'AXIE': 'AXS',
    'USDT': 'USDT',
    'ATOM': 'ATOM',
    'NEAR': 'NEAR',
    'AVAX': 'AVAX',
    'GLMR': 'GLMR',
    'DOGE': 'DOGE',
    'BUSD': 'BUSD',
}
```

```python
df_values
```

```python
df_values['Símbolo'] = df_values['Ativo'].map(dict_transform)
# df_values['Preço Médio'] = np.nan
```

```python
df_values
```

```python
quantity = df_values.groupby(['Data', 'Símbolo']).sum().drop(columns='Total')
```

```python
quantity
```

```python
btc_qtt = quantity[quantity.index.get_level_values(1) == 'BTC']
hoje_datetime = datetime.datetime.now()
```

```python
index = pd.date_range(start = df_values['Data'].min(), end = hoje_datetime, freq='D')
df = pd.DataFrame(index=index)
```

```python
btc_qtt = btc_qtt.reset_index().set_index('Data')
```

```python
btc_qtt['Quantidade Acumulada'] = btc_qtt['Quantidade'].cumsum()
```

```python
btc_qtt = btc_qtt.drop(columns=['Quantidade']).resample('D').ffill()
```

```python
btc_price = yf.Ticker(btc_qtt['Símbolo'].unique()[0] + '-USD').history(start = btc_qtt.index[1], 
                                                           end = hoje_datetime)['Close']
brlusd = yf.Ticker('BRL=X').history(start = btc_qtt.index[1], 
                                    end = hoje_datetime)['Close']
```

```python
btc_qtt['Preço'] = btc_price 
btc_qtt['Câmbio'] = brlusd
btc_qtt['Câmbio'].ffill(inplace=True)
btc_qtt['Preço'] = btc_price * btc_qtt['Câmbio']
```

```python
btc_qtt['Patrimônio'] = btc_qtt['Quantidade Acumulada'] * btc_qtt['Preço']
```

```python
btc_qtt['Patrimônio'].plot(figsize=(19,5))
hoje = btc_qtt['Patrimônio'][-1]

print(f'Patrimônio de BTC hoje: R${round(hoje, 2)}')
```
